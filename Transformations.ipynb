* Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)


```python
# Vinculamos google colab con google drive para importar el dataset

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Importamos librerias

import pandas as pd
```

```python
# Levantamos los datasets, guardamos cada uno en una variable distinta y agregamos la columna Id

data_amazon = pd.read_csv('/content/drive/MyDrive/MyLops/amazon_prime_titles-score.csv', sep=',', encoding='latin-1')
data_amazon['Id'] = 'a' + data_amazon['show_id']

data_disney = pd.read_csv('/content/drive/MyDrive/MyLops/disney_plus_titles-score.csv', sep=',', encoding='latin-1')
data_disney['Id'] = 'd' + data_disney['show_id']

data_hulu = pd.read_csv('/content/drive/MyDrive/MyLops/hulu_titles-score (2).csv', sep=',', encoding='latin-1')
data_hulu['Id'] = 'h' + data_hulu['show_id']

data_netflix = pd.read_csv('/content/drive/MyDrive/MyLops/netflix_titles-score.csv', sep=',', encoding='latin-1')
data_netflix['Id'] = 'n' + data_netflix['show_id']
```

```python
# Antes de unirlos, verificamos que los dataset tengan igual cant de columnas
# Tambien ver cant de filas, para luego sumarlas y verificar que todas se agregaron al dataset total

print(data_amazon.shape)
print(data_disney.shape)
print(data_hulu.shape)
print(data_netflix.shape)
```

```python
# Unimos los dataset en uno solo

data_plataformas = pd.concat([data_amazon, data_disney, data_hulu, data_netflix])
data_plataformas.head()
```

```python
# Cant de registros y columnas del dataset de las plataformas

data_plataformas.shape
```

### **``Transformación 2:``**
Los valores nulos del campo rating deberán reemplazarse por el string “**`G`**” (corresponde al maturity rating: “general for all audiences”

```python
# Verificamos cuantos valores nulos hay por cada columna

data_plataformas.isna().sum()
```

```python
# Reemplazamos los valores nulos de la columna Rating por G

data_plataformas['rating'].fillna('g', inplace=True)
data_plataformas.head()
```

```python
# Verificamos que la columna rating no cuenta con valores nulos luego de reemplazarlos.

data_plataformas['rating'].isna().sum()
```

### **``Transformación 3:``**
De haber fechas, deberán tener el formato **`AAAA-mm-dd`**

```python
# Cambiamos el formato fecha a aaaa-mm-dd

data_plataformas['date_added'] = pd.to_datetime(data_plataformas['date_added'])
data_plataformas.head()
```

### **``Transformación 4:``**
Los campos de texto deberán estar en **minúsculas**, sin excepciones

```python
# Todos los campos textos pasan a minusculas

data_plataformas['type'] = data_plataformas['type'].str.lower()
data_plataformas['title'] = data_plataformas['title'].str.lower()
data_plataformas['director'] = data_plataformas['director'].str.lower()
data_plataformas['cast'] = data_plataformas['cast'].str.lower()
data_plataformas['country'] = data_plataformas['country'].str.lower()
data_plataformas['rating'] = data_plataformas['rating'].str.lower()
data_plataformas['duration'] = data_plataformas['duration'].str.lower()
data_plataformas['listed_in'] = data_plataformas['listed_in'].str.lower()
data_plataformas['description'] = data_plataformas['description'].str.lower()
data_plataformas['Id'] = data_plataformas['Id'].str.lower()
data_plataformas.head()
```

### **``Transformación 5:``**
El campo ***duration*** debe convertirse en dos campos: **`duration_int`** y **`duration_type`**. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
# Extraemos la cant de la columna duración

data_plataformas['duration_int'] = data_plataformas['duration'].str.extract('([0-9]+)')
data_plataformas.head()
```

```python
# Extraemos el tipo de duración de la columna duración

data_plataformas['duration_type'] = data_plataformas['duration'].str.extract('([a-z]+)')
data_plataformas.head()
```

```python
# Eliminamos las columnas sobrantes

data_plataformas.drop(['show_id'], axis=1, inplace=True)
data_plataformas.drop(['duration'], axis=1, inplace=True)
data_plataformas.head()
```

```python
# Verificamos que campos tienen valores nulos

data_plataformas.isna().sum()
```

```python
# Completamos los valores nulos del campo "duration_int" con 0

data_plataformas['duration_int'].fillna(0, inplace=True)
data_plataformas['duration_int'].isna().sum()
```

```python
# Llenamos los campos vacios con tipo de dato string con la leyenda "sin dato"

data_plataformas['director'].fillna('sin dato', inplace=True)
data_plataformas['cast'].fillna('sin dato', inplace=True)
data_plataformas['country'].fillna('sin dato', inplace=True)
data_plataformas['duration_type'].fillna('sin dato', inplace=True)
data_plataformas
```

```python
# Verificamos si en duration_type, que aparezcan valores unicos.

data_plataformas_type = data_plataformas['duration_type'].unique()
data_plataformas_type
```

```python
# Unificamos los valores de seasons

data_plataformas['duration_type'].replace({'season': 'seasons'}, inplace=True)
data_plataformas
```

```python
# Verificamos que los valores únicos del campo 'duration_time' sean correctos

data_plataformas_type = data_plataformas['duration_type'].unique()
data_plataformas_type
```

```python
# Verificamos que no hayan quedado valores nulos, salvo los campos con formato fecha (date)

data_plataformas.isna().sum()
```

```python
# Levantamos cada csv con los rating de las peliculas

rating1 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/1.csv', sep=',', encoding='latin-1')
rating2 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/2.csv', sep=',', encoding='latin-1')
rating3 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/3.csv', sep=',', encoding='latin-1')
rating4 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/4.csv', sep=',', encoding='latin-1')
rating5 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/5.csv', sep=',', encoding='latin-1')
rating6 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/6.csv', sep=',', encoding='latin-1')
rating7 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/7.csv', sep=',', encoding='latin-1')
rating8 = pd.read_csv('/content/drive/MyDrive/MyLops/ratings/8.csv', sep=',', encoding='latin-1')

print(rating1.shape)
print(rating2.shape)
print(rating3.shape)
print(rating4.shape)
print(rating5.shape)
print(rating6.shape)
print(rating7.shape)
print(rating8.shape)
```

```python
# Unimos los dataset de ratings

data_score = pd.concat([rating1, rating2, rating3, rating4, rating5, rating6, rating7, rating8])
data_score.shape
```

```python
# Verificamos si tienen valores nulos o faltantes

data_score.isna().sum()
```

```python
# Creamos un nuevo dataset para agrupar el rating promedio por pelicula

data_score_prom = data_score.groupby(['movieId']).mean()
data_score_prom
```

```python
# Unimos ambos dataset, mientras que el id de cada uno coincida

data_plataformas_prom = pd.merge(left=data_plataformas, right=data_score_prom, how='left', left_on='Id', right_on='movieId')
data_plataformas_prom.head()
```

```python
# Eliminamos las columnas que nos sobran (userId y timestamp)

data_plataformas_prom.drop(['userId'], axis=1, inplace=True)
data_plataformas_prom.drop(['timestamp'], axis=1, inplace=True)
data_plataformas_prom
```

```python
# Cambiamos el nombre de la columna rating por score

data_plataformas_prom.rename(columns={'rating_y':'score'}, inplace=True)
data_plataformas_prom
```

```python
# Redondeamos los campos float a 1 solo decimal

data_plataformas_prom = data_plataformas_prom.round(1)
data_plataformas_prom
```

```python
# Info general del dataset final

data_plataformas_prom.info()
```

```python
# Generamos el csv que trabajamos con los datasets de las plataformas y el score promedio para usarlo en la Api

data_plataformas_prom.to_csv('/content/drive/MyDrive/MyLops/plataformas_prom.csv', index=False)
```

```python
# Ahora debemos dejar listo el dataset con los score, sin filtrar datos, con el formato correcto

data_score.head()
```

```python
# Valores nulos ya habiamos verificado y no tiene.

# Ahora pasamos a cambiar el formato fecha unix a formato fecha aaaa-mm-dd

data_score['timestamp'] = pd.to_datetime(data_score['timestamp'], unit='s').dt.strftime('%Y-%m-%d')
data_score
```
